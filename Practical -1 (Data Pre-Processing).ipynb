# Data Pre-Processing

```python
#importing necessary libraries
import pandas as pd
import numpy as np
```

```python
#importing dataset (placement.csv)
data = pd.read_csv('placement 1.csv')
data
```

```python
#checking and handling the missing values
data.isnull().sum().sum()
```

```python
#Unnamed:0 column is not needed therefore we will drop it
data.drop(data.columns[0], axis = 1, inplace = True)
data
```

```python
#dropping duplicate rows
data.drop_duplicates()
```

```python
#handling missing values
#handling numerical missing values by replacing null values with mean of the entire column
#using sklearn.impute library's meathod (SimpleImputer)

from sklearn.impute import SimpleImputer as si

imputer = si(strategy = 'mean', missing_values = np.nan)
imputer = imputer.fit(data[['degree_p','etest_p','mba_p','salary']])
data[['degree_p','etest_p','mba_p','salary']] = imputer.transform(data[['degree_p','etest_p','mba_p','salary']])
data
```

```python
#checking null values in numerical columns
data[['degree_p','etest_p','mba_p','salary']].isnull().sum()
```

```python
#handling categorical missing values
# checking gender column's null values and droping rows which have null values.
data[['gender']].isnull().sum()
```

```python
data['gender'].dropna(how = 'any', inplace = True)
data[['gender']].isnull().sum()
```

```python
#replacing rest of categorical column's null values with most-repeated value using sklearn.impute library's method
#SimlpeImputer

from sklearn.impute import SimpleImputer as si

imputer = si(strategy = 'most_frequent', missing_values = np.nan)
imputer = imputer.fit(data[['degree_t','workex','specialisation']])
data[['degree_t','workex','specialisation']] = imputer.transform(data[['degree_t','workex','specialisation']])
data
```

```python
#all missing values resolved
data.isnull().sum()
```

# Encoding Categorical Data

```python
#splitting data in independent and dependent variables
x = data[['gender','degree_t','workex','specialisation']]
y = data[['salary']]
```

```python
#encoding categorical using label encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
x.loc[:,'gender'] = le.fit_transform(x.loc[:,'gender'])
x.loc[:,'workex'] = le.fit_transform(x.loc[:,'workex'])
x.loc[:,'specialisation'] = le.fit_transform(x.loc[:,'specialisation'])
x
```

```python
x['degree_t'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(),[1])], remainder = 'passthrough')
x = ct.fit_transform(x)
x
```

```python
x = pd.DataFrame(x, columns = ['Sci&Tech','Comm&Mgmt','Others','gender','workex','specialisation'])
x
```

```python
#splitting data into training and test dataset
from sklearn.model_selection import train_test_split as tts
x_train ,x_test, y_train, y_test = tts(x,y, test_size =0.2)
```
